# Conceptual Models Explaination:

That Represents Consumers,Producers and Companies in our domain:

Conceptual 2.png

The domain is TV shows and its representation is broadly divided into three main categories: Consumers, Producers and Companies. The Consumers for the TV Shows are fans i.e. the audience that watch the respective shows. The Producers are the executive producers or any one of the many producers the TV Show has. Finally, the Companies in our domain are the companies which own the right to air the shows or the companies which originally aired the show.

For any user in our domain, they have their own parameters such as 'Title','channelTitle','commentCount','likeCount','dislikeCount','tags','videoId','categordyId','publishedAt' i.e. timestamp of the videos. Also, they have their Youtube Channel's channelId too, which is infact unique and we are using that as our Primary Key for the Consumer table.

The Consumer has various parameters, but they can be identified uniquely by ‘channelId’ and each consumer i.e. the fan of the TV Show will have their Primary key as channelId and that will not match with any other fan’s channelId.

The Producer and Companies also have various parameters, Primary keys for both are ‘User_id’ in the respective databases. These Primary Keys are the best way to make relations with multiple tables because they are unique in their own sense.

# Entity Relationship Diagram:

ER.png

As explained in the Conceptual Models, the Shows have - Companies,Producers and Consumers and each of them have their own database. The Relation betweeen them is multifarious.

The 'Shows' table have many to many and one to many relations with other threee tables. The 'Shows' table has one to many relation with 'Consumer' table, from 'Show' table we can get the Company of the show and from 'Consumer' table we can get that show's attributes such as channel Title, channel ID- used as a primary key, Like count, comment count and all the column names of 'Consumer' can relate to Company from 'Shows'.

Also, the 'Shows' table has many to many relationship with 'Producer' table. Producer's name and Producer's screen name from 'Shows' table can be realted with many attributes of 'Producer' table like User ID - which is used as a primary key,Name of the Show, Producer's Tweet, Followeres and all the attributes from that table.

Similarly, the 'Shows' table also has many to many relationship with 'Company' Table. Company's name and Company's Screen name can be related to Company's user ID which is a primary key and other attributes.

The 'Consumer' table has one to one relationship with 'Producer' table. We can get the Channel Title from 'Consumer' and get the producer for the show from 'Producer' table.

The 'Consumer' table also has one to one relationship with 'Company' table. We can relate the Channel Title from 'Consumer' withe the Company's name from 'Company' table.

2)When should you use an entity versus attribute?

An entity is the set of objects of which the data is to be collected and stored in a database. An entity can be used to participate in multiple relations and they have attributes which are the properties of the entities. These attributes are the properties which can’t be defined independently without defining its entity.

Entity: An object for which data is to be collected and processed.
Relationship: Which related two entities either by their attributes or relating entities directly
